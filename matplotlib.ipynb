```python
#Read tables
table = pd.read_csv('C:/Users/daani/Documents/Python codes/redshift1.csv')
table
```

```python
#Load rows of tables as variables
x = table.distance
y = table.z	
```

```python
#Scatter Plot
plt.xlabel('distance')
plt.ylabel('redshift')
plt.scatter(y,x)
plt.show()
```

```python
#Relation based on graph
rand1 = int((table.distance[0])/(table.z[0]))
rand2 = int((table.distance[99])/(table.z[99]))
if rand1 == rand2:
    slope = (True)
else:
    slope = (False)
print("redshift and distance is related:",slope)
```
